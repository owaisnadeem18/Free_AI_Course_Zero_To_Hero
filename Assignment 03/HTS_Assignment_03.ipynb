```python

# Task : 04

# Write a function called "is_even" that takes a number as an argument and returns True if the number is even, and False otherwise

#Solution of Task : 04

def is_even(number): # Here I created a function named { is_even(parameter) } # number is a parameter here

    if number % 2 == 0: # now I'm saying if number is getting exactly divisible by 2 then it means that it is an even number so just return true

        return True

    else:

        return False # otherwise return false if number is not getting completely divided by 2 (which means that number is odd)

print(is_even (6)) # 06 is the argument here 
```
